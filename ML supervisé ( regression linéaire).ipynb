```python
import numpy as np
import matplotlib.pyplot as plt
```

Regression with SKLearn

```python
m=100 # creating 100 samples
x=np.linspace(0,10,m).reshape(m,1) # important declaration de la dim les tableaux numpy x et y ont 2 dimensions
y=x+np.random.randn(m,1)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
model.score(x,y)
```

```python
model = LinearRegression()
model.fit(x,y)
model.score(x,y)
prediction = model.predict(x)

plt.scatter(x,y)
plt.plot(x, prediction, c='r')
```

```python
m=100 # creating 100 samples
x=np.linspace(0,10,m).reshape(m,1) # important declaration de la dim les tableaux numpy x et y ont 2 dimensions
y=x**2+np.random.randn(m,1)
```

```python
model = LinearRegression()
model.fit(x,y)
model.score(x,y)
prediction = model.predict(x)

plt.scatter(x,y)
plt.plot(x, prediction, c='r')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
model = SVR(C=100)
model.fit(x,y)
model.score(x,y)
prediction = model.predict(x)

plt.scatter(x,y)
plt.plot(x, prediction, c='r')
```
