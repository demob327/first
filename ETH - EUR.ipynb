```python
#plt.figure(figsize(12,8))
ethereum.loc['2019','Close'].plot()
ethereum.loc['2019','Close'].resample('M').mean().plot(label= 'Moyenne par mois', lw=3, ls=':', alpha=0.8)
ethereum.loc['2019','Close'].resample('W').mean().plot(label= 'Moyenne par semaine', lw=2, ls='--', alpha=0.8)
plt.legend()
plt.show()
```

```python
ethereum.loc['2019','Close'].resample('W').agg(['mean', 'std', 'min', 'max']).head(8)
```

```python
ethereum.loc['2019','Close'].resample('W').agg(['mean', 'std', 'min', 'max']).plot(subplots = True)
```

```python
m=ethereum['Close'].resample('W').agg(['mean', 'std', 'min', 'max'])

plt.figure(figsize=(12,8))
m['mean']['2019'].plot(label= 'moyenne par semaine')
plt.fill_between(m.index, m['max'], m['min'], alpha=0.3, label= 'min - max par semaine')

plt.legend()
plt.show()
```
