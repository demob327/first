```python
import numpy as np
from sympy import symbols, diff, solve

# Définition de la fonction g(x, y)
def g(x, y):
    return x**2 * y + y**2 * x - x * y

# Calcul des dérivées partielles de g(x, y)
x, y = symbols('x y')
dg_dx = diff(g(x, y), x)
dg_dy = diff(g(x, y), y)

# Résolution des équations dg_dx = 0 et dg_dy = 0
critical_points = solve((dg_dx, dg_dy), (x, y))

# Affichage des résultats
print("Points critiques :")
for point in critical_points:
    print(f"Point : {point}")
```

```python
def find_solution(target):
    for a in range(20):  # Recherche des valeurs jusqu'à a = 19
        for b in range(a+1, 20):  # b doit être strictement supérieur à a
            for c in range(b+1, 20):  # c doit être strictement supérieur à b
                if 2**a + 2**b + 2**c == target:
                    return a, b, c
    return None  # Si aucune solution n'est trouvée

target = 1120
solution = find_solution(target)

if solution:
    a, b, c = solution
    print(f"Valeurs trouvées : a = {a}, b = {b}, c = {c}")
else:
    print("Aucune solution trouvée.")
```

```python
import numpy as np
from fractions import Fraction

def numpy_frac(value):
    if isinstance(value, (int, np.int64, float)):
        frac = Fraction(value).limit_denominator()
        numerator = frac.numerator
        denominator = frac.denominator
        return f"{numerator}/{denominator}"
    else:
        return str(value)

def calculer_valeurs_propres(matrice):
    valeurs_propres, vecteurs_propres = np.linalg.eig(matrice)
    return valeurs_propres, vecteurs_propres

# Exemple d'utilisation pour une matrice d'ordre 7
n = 7  # Ordre de la matrice
matrice = np.random.rand(n, n)  # Remplacez cette ligne par votre propre matrice

valeurs_propres, vecteurs_propres = calculer_valeurs_propres(matrice)

print("Valeurs propres :")
for valeur in valeurs_propres:
    valeur_fraction = numpy_frac(valeur)
    print(valeur_fraction)

print("\nVecteurs propres :")
print(vecteurs_propres)
```
