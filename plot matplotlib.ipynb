```python
import matplotlib.patches as mpatches
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.style.use('ggplot')
mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std) 
plt.minorticks_on()
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='y', colors='g', grid_color='gray', grid_alpha=0.5)
PDF= mpatches.Patch(color='b', label='PDF')
plt.legend(handles=[PDF])
plt.plot(x, y,'c--',lw=3)
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
plt.rcParams['figure.figsize'] = [10/2.54, 8/2.54]
plt.show()
```
