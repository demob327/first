```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Fonction objectif
def f(x, y):
    return x**2 * np.exp(y) * y

# Contrainte
def g(x, y):
    return x**2 + y**2 - 4

# Contrainte galit
def constraint_eq(x):
    return g(x[0], x[1])

# Fonction objectif  maximiser (ngative de la fonction d'origine)
def objective(x):
    return -f(x[0], x[1])

# Valeurs initiales
x0 = [0, 0]

# Rsolution du problme d'optimisation
solution = minimize(objective, x0, constraints={'type': 'eq', 'fun': constraint_eq})

# Cration de la grille de valeurs
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Calcul de la fonction objectif sur la grille
Z = f(X, Y)

# Calcul de la contrainte sur la grille
C = g(X, Y)

# Affichage graphique
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Surface de la fonction objectif
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Contrainte (cercle de rayon 2)
theta = np.linspace(0, 2*np.pi, 100)
x_circle = 2 * np.cos(theta)
y_circle = 2 * np.sin(theta)
ax.plot(x_circle, y_circle, np.zeros_like(x_circle), color='r')

# Point optimal (maximum)
ax.scatter(solution.x[0], solution.x[1], -solution.fun, color='b', label='Maximum')

# Labels des axes
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

# Lgende
ax.legend()

# Affichage du graphique
plt.show()
```

```python
import numpy as np

# Fonction objectif
def f(x, y):
    return x**2 * np.exp(y) * y

# Contrainte
def g(x, y):
    return x**2 + y**2 - 4

# Discrtisation de la grille de valeurs
n = 100  # Nombre de points sur chaque axe
x = np.linspace(-2, 2, n)
y = np.linspace(-2, 2, n)

# Initialisation du maximum
max_value = -np.inf
max_x = None
max_y = None

# Recherche du maximum
for i in range(n):
    for j in range(n):
        if g(x[i], y[j]) == 0:  # Vrification de la contrainte
            value = f(x[i], y[j])
            if value > max_value:  # Mise  jour du maximum
                max_value = value
                max_x = x[i]
                max_y = y[j]

# Affichage des rsultats
if max_x is not None and max_y is not None:
    print("Le maximum de la fonction est :", max_value)
    print("Obtenu pour x =", max_x)
    print("et y =", max_y)
else:
    print("Aucune solution trouve dans la plage spcifie.")

```

```python
import numpy as np

def exponentiation_matricielle(A, t):
    # Diagonalisation de la matrice A
    eigenvalues, eigenvectors = np.linalg.eig(A)
    P = eigenvectors
    P_inv = np.linalg.inv(P)
    D = np.diag(eigenvalues)

    # Calcul de la matrice exponentielle
    exp_D = np.diag(np.exp(np.diagonal(D) * t))
    exp_A = P @ exp_D @ P_inv

    return exp_A

# Matrice A
A = np.array([[2, 0, 1], [1, -1, -1], [-1, 2, 2]])

# Valeur de t
t = 1

# Calcul de e^(A*t)
resultat = exponentiation_matricielle(A, t)
print(resultat)
```
