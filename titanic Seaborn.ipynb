```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(5)
```

```python
titanic.drop(['class', 'who', 'adult_male', 'embark_town', 'alive', 'alone'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

pour amliorer la visualisation on utilise les fonctions de categories car on a plusieures catgories dans le dataset

```python
sns.catplot(x = 'pclass', y = 'age', data = titanic)
```

```python
sns.catplot(x = 'pclass', y = 'age', data = titanic, hue = 'sex')
```

```python
sns.boxplot(x = 'pclass', y = 'age', data = titanic, hue = 'sex')
```

Visualiser la distribution du dataset selon le tarif (col = fare)

```python
sns.distplot(titanic['fare'])
```

on voit que le tarif ne suit pas une loi normale ; la majorit ayant opt pour le tarif le - cher.

```python
sns.jointplot('age', 'fare', data =titanic)
```

```python
sns.jointplot('age', 'fare', data =titanic, kind = 'kde') # kerne density estimation
```

```python
sns.jointplot('age', 'fare', data =titanic, kind = 'hex') # hexagonal
```

```python
sns.heatmap(titanic.corr())
```
